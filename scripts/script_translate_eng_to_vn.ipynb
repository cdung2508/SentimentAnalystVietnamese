```python
import pandas as pd
import numpy as np
import translators as ts


def translate_to_vietnamese(text):
    return ts.google(text, to_language='vi')


def filter_function(x):
    if x == 'No Negative' or x == 'No Positive':
        return ''
    return x


df = pd.read_csv('/content/drive/MyDrive/Hotel_Reviews_Raw_Eng.csv')
data = df[['Negative_Review', 'Positive_Review']]
del df

negative_review = data['Negative_Review'].apply(
    lambda x: filter_function(x)).values
positive_review = data['Positive_Review'].apply(
    lambda x: filter_function(x)).values
del data

negative_review = [review for review in negative_review.tolist(
) if review != '' or review == None]
positive_review = [review for review in positive_review.tolist(
) if review != '' or review == None]
label_negative = [0 for _ in range(len(negative_review))]
label_positive = [1 for _ in range(len(positive_review))]

data = np.array(negative_review[27000:28000] +
                positive_review[27000:28000]).reshape(-1, 1)
label = np.array(label_negative[27000:28000] +
                 label_positive[27000:28000]).reshape(-1, 1)
del label_negative
del label_positive
del negative_review
del positive_review

current_index = 0
for text in np.nditer(data):
    try:
        data[current_index] = translate_to_vietnamese(text.item())
        print(
            f'Text after translate in index {current_index}: "{data[current_index].item()}"')
        print('-' * 40)
        current_index += 1
    except:
        print('Error in index:', current_index)
        current_index += 1
        continue
```

```python
dataset = np.c_[data, label]
```

```python
dataset = pd.DataFrame(dataset, columns=['Word', 'Label'])
```

```python
filepath = 'review_vietnamese_dataset_28.csv'
```

```python
dataset.to_csv(filepath, index=False)
```
